```python
import pandas as pd

f3 = pd.read_csv('like_post.csv')

f3 = f3.drop(columns = ['comment_id'])
f3
```

```python
f4 = f3.groupby(['user_id','friend_id','response']).count().drop(columns = ['user_name','friend_name'])
f4

f4.reset_index().to_csv('groupby2.csv',index=False)
```

```python
f4 = pd.read_csv('groupby2.csv')

l4 = len(f4)
#print(l4)

W = []

for i in range(l4):
    if f4.response[i] == 'LIKE':
        W.append(f4.post[i] * 0.25)
    elif f4.response[i] == 'LOVE':
        W.append(f4.post[i] * 0.75)
    elif f4.response[i] == 'WOW':
        W.append(f4.post[i] * 0.5)
    elif f4.response[i] == 'SAD':
        W.append(f4.post[i] * (-0.5))
    elif f4.response[i] == 'ANGRY':
        W.append(f4.post[i] * (-1))  
    elif f4.response[i] == 'HAHA':
        W.append(f4.post[i] * 0.25)
    elif f4.response[i] == 'THANKFUL':
        W.append(f4.post[i] * 0.25)

#for i in range(l4):
 #   print(W[i])
    
f4['weights'] = W 
f4.to_csv('user_weight_likes.csv', index = False)
```
