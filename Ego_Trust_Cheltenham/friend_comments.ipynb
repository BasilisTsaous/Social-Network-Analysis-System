```python
import pandas as pd

f1 = pd.read_csv('new_posts1-2.csv')
f2 = pd.read_csv('new_comments.csv')
l1 = len(f1)
l2 = len(f2)
#print(l1,l2)

user_id = []
post = []
user_name = []

comment_id = []
friend_id = []
friend_name = []
rid = []
msg = []

c = 0 

for i in range(l1):
    for k in range(l2):
        if f1.pid[i] == f2.pid[k] :
            c = c + 1
            
            user_name.append(f1.name[i])
            user_id.append(f1.id[i])
            post.append(f1.pid[i])
            
            comment_id.append(f2.cid[k])
            friend_name.append(f2.name[k])
            friend_id.append(f2.id[k])
            rid.append(f2.rid[k])
            msg.append(f2.msg[k])
            
            
#print(c) 

dict = {'user_name' : user_name, 'user_id': user_id, 'post': post, 'comment_id': comment_id, 'friend_name': friend_name, 'friend_id': friend_id, 'rid' : rid, 'msg' : msg} 
df = pd.DataFrame(dict)
df.to_csv('friend_comments.csv', index = False)

#for i in range(c):
 #   print(user[i],post[i],comment_id[i],friend_id[i],rid[i],msg[i],sep=',')
```
