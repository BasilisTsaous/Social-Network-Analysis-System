```python
import csv
import numpy as np

# open the file in universal line ending mode 
with open('like.csv', 'r') as like_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(like_file)
  like = {}
  for row in reader:
    for header, value in row.items():
      try:
        like[header].append(value)
      except KeyError:
        like[header] = [value]

#Put all the parameters in separate lists
gid = like['gid']
pid = like['pid']  
cid = like['cid']
response = like['response']
uid = like['id']
name = like['name']    

#print(response)
l_response = len(response)  # the number of the post responses in the network

count_post = 0

Post_id = [] 
Posts = []


R_LIKE = []
R_SAD = []
R_ANGRY = []
R_LOVE = []
R_WOW = []
R_HAHA = []
R_THANKFUL = []

for i in range(l_response): 
    # Initialization of the 7 counters
    r1=0
    r2=0
    r3=0
    r4=0
    r5=0
    r6=0
    r7=0
    
    same_post=1  
    # 7 conditions, to find the response of the posts in the network
    if response[i] == 'LIKE' or response[i] == 'LIKES':
        r1=r1+1  
    elif response[i] == 'SAD':
        r2=r2+1
    elif response[i] == 'ANGRY':
        r3=r3+1
    elif response[i] == 'LOVE':
        r4=r4+1
    elif response[i] == 'WOW':
        r5=r5+1
    elif response[i] == 'HAHA':
        r6=r6+1
    elif response[i] == 'THANKFUL':
        r7=r7+1
        
    for k in range(i+1,l_response):
        if pid[i] == pid[k] and i != k:   #Search for the different post id's in the file
            same_post = same_post + 1  #Increases every time we find a cell with the same pid(post id)
            pid[k] = ''
            
            if response[k] == 'LIKE' or response[k] == 'LIKES':
                r1=r1+1   #Increases every time we find a cell with response = LIKE
            elif response[k] == 'SAD':
                r2=r2+1  #Increases every time we find a cell with response = SAD
            elif response[k] == 'ANGRY':
                r3=r3+1  #Increases every time we find a cell with response = ANGRY
            elif response[k] == 'LOVE':
                r4=r4+1  #Increases every time we find a cell with response = LOVE
            elif response[k] == 'WOW':
                r5=r5+1  #Increases every time we find a cell with response = WOW
            elif response[k] == 'HAHA':
                r6=r6+1  #Increases every time we find a cell with response = HAHA
            elif response[k] == 'THANKFUL':
                r7=r7+1  #Increases every time we find a cell with response = THANKFUL
                
    
    if pid[i] != '':   
        count_post = count_post + 1  #Denotes the number of different posts in the network
        Post_id.append(pid[i])  #Append into the list 'Post_id', the id's of the different posts(pid) we found
        Posts.append(same_post) #Append into the list 'Posts', the number of the posts for each post id
        
        R_LIKE.append(r1)  #Append into the list 'R_LIKE', the total number of 'LIKE' responses for each post
        R_SAD.append(r2)   #Append into the list 'R_SAD', the total number of 'SAD' responses for each post
        R_ANGRY.append(r3) #Append into the list 'R_ANGRY', the total number of 'ANGRY' responses for each post
        R_LOVE.append(r4)  #Append into the list 'R_LOVE', the total number of 'LOVE' responses for each post
        R_WOW.append(r5)   #Append into the list 'R_WOW', the total number of 'WOW' responses for each post
        R_HAHA.append(r6)  #Append into the list 'R_HAHA', the total number of 'HAHA' responses for each post
        R_THANKFUL.append(r7) #Append into the list 'R_THANKFUL', the total number of 'THANKFUL' responses for each post

#print(count_post)        
#for i in range(count_post):
 #   print(Post_id[i],R_LIKE[i],R_SAD[i],R_ANGRY[i],R_LOVE[i],R_WOW[i],R_HAHA[i],R_THANKFUL[i], sep = ',')
    

with open('post_response.csv', mode='w') as post_response_file:
    fieldnames = ['post_id', 'post_likes','post_sad','post_angry','post_love','post_wow','post_haha','post_thankful']
    writer = csv.DictWriter(post_response_file, fieldnames=fieldnames)

    writer.writeheader()
    for i in range(count_post):
        writer.writerow({'post_id': Post_id[i], 'post_likes': R_LIKE[i] ,'post_sad': R_SAD[i] , 'post_angry': R_ANGRY[i], 'post_love': R_LOVE[i], 'post_wow': R_WOW[i], 'post_haha': R_HAHA[i], 'post_thankful': R_THANKFUL[i]})
        #print(Post_id[i],R_LIKE[i],R_SAD[i],R_ANGRY[i],R_LOVE[i],R_WOW[i],R_HAHA[i],R_THANKFUL[i], sep = ',')
```
