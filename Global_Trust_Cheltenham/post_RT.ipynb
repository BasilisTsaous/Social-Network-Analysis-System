```python
import csv
import numpy as np

# open the file in universal line ending mode 
with open('final_post.csv', 'r') as final_post_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(final_post_file)
  f_post = {}
  for row in reader:
    for header, value in row.items():
      try:
        f_post[header].append(value)
      except KeyError:
        f_post[header] = [value]
        
#Put all the parameters in separate lists
p_id = f_post['post_id']
p_comments = f_post['post_comments']  
p_shares = f_post['post_shares']

l_pid = len(p_id)  # the number of the id's(posts) in the network


# open the file in universal line ending mode 
with open('post_weight_response.csv', 'r') as pw_response_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(pw_response_file)
  pw = {}
  for row in reader:
    for header, value in row.items():
      try:
        pw[header].append(value)
      except KeyError:
        pw[header] = [value]

#Put all the parameters in separate lists       
post_id = pw['post_id']
post_likes = pw['post_likes']
post_sad = pw['post_sad']
post_angry = pw['post_angry']
post_love = pw['post_love']
post_wow = pw['post_wow']
post_haha = pw['post_haha']
post_thankful = pw['post_thankful']
post_weight = pw['post_weight']

l_post = len(post_id)  # the number of the post id's in the network


#Convert the lists of the file into numeric arrays for calculations
P_COMMENTS = np.asarray(p_comments, dtype = int)
P_SHARES = np.asarray(p_shares, dtype = int)

P_WEIGHT = np.asarray(post_weight, dtype = float)

P_LIKES = np.asarray(post_likes, dtype = int)
P_SAD = np.asarray(post_sad, dtype = int)
P_LOVE = np.asarray(post_love, dtype = int)
P_ANGRY = np.asarray(post_angry, dtype = int)
P_WOW = np.asarray(post_wow, dtype = int)
P_HAHA = np.asarray(post_haha, dtype = int)
P_THANKFUL = np.asarray(post_thankful, dtype = int)


                  # RESPONSE TRUST

S1 = []
S2 = []
POST = []

c = 0

for i in range(l_pid):
    s1 = 0
    s2 = 0
    for k in range(l_post):
        if p_id[i] == post_id[k]:  #Search for the posts(post id's) that exist in both files
            s1 = P_COMMENTS[i] * 0.75 + P_SHARES[i] * 1 + P_WEIGHT[k]  #Calculate the total (action*weight) for each post
            s2 = P_COMMENTS[i] + P_SHARES[i] + P_LIKES[k] + P_SAD[k] + P_ANGRY[k] + P_LOVE[k] + P_WOW[k] + P_HAHA[k] + P_THANKFUL[k] #Calculate the total number of actions for each post
            c = c + 1  #Increases every time we find the same post id in both files

            S1.append(s1) #Append into the 'S1' list, the total weight we found for each post
            S2.append(s2) #Append into the 'S2' list, the total number of actions we found for each post
            POST.append(p_id[i]) #Append into the 'POST' list, the id's of the different posts we found 
    
ResT = [] 
POST_WEIGHT = np.asarray(S1, dtype = float)  ##Convert the list 'S1' into the numeric array 'POST_WEIGHT' for calculations 
POST_ACT = np.asarray(S2, dtype= float)  ##Convert the list 'S2' into the numeric array 'POST_ACT' for calculations 

s3 = []
S3 = []

for i in range(c):
    s3 = POST_WEIGHT[i] / POST_ACT[i]  #Calculate the RESPONSE TRUST for each post we found and store the result in the 's3' list
    S3 = round(s3,4)  #Store in the new list 'S3' , the result we found, rounded to 2 decimal places
    ResT.append(S3)  #Append in the list 'ResT', the final result 'S3' for the RESPONSE TRUST of each post
    
#for i in range(l_pid):
 #   print(POST[i],ResT[i],sep = ',')

with open('post_RT.csv', mode='w') as post_RT_file:
    fieldnames = ['post_id', 'Response Trust']
    writer = csv.DictWriter(post_RT_file, fieldnames=fieldnames)

    writer.writeheader()
    for i in range(c):
        writer.writerow({'post_id': POST[i], 'Response Trust': ResT[i]})
        #print(POST[i],ResT[i],sep = ',')  
```
