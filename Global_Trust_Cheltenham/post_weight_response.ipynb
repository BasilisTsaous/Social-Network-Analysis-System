```python
import csv
import numpy as np

# open the file in universal line ending mode 
with open('post_response.csv', 'r') as post_response_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(post_response_file)
  pr_like = {}
  for row in reader:
    for header, value in row.items():
      try:
        pr_like[header].append(value)
      except KeyError:
        pr_like[header] = [value]

#Put all the parameters in separate lists     
post_id = pr_like['post_id']
post_likes = pr_like['post_likes']
post_sad = pr_like['post_sad']
post_angry = pr_like['post_angry']
post_love = pr_like['post_love']
post_wow = pr_like['post_wow']
post_haha = pr_like['post_haha']
post_thankful = pr_like['post_thankful']

l_post = len(post_id)  # the number of the post id's in the network


W = []
WEIGHT = []

#Convert the lists of the parameters into numeric arrays, for operations(calculations) 
P_LIKE = np.asarray(post_likes, dtype=int)
P_SAD = np.asarray(post_sad, dtype=int)
P_ANGRY = np.asarray(post_angry, dtype=int)
P_LOVE = np.asarray(post_love, dtype=int)
P_WOW = np.asarray(post_wow, dtype=int)
P_HAHA = np.asarray(post_haha, dtype=int)
P_THANKFUL = np.asarray(post_thankful, dtype=int)


for i in range(l_post):
    # Calculate the combined weight of the 7 responses of each post and store it in the list 'W'
    W = P_LIKE[i] * 0.25 - P_SAD[i] * 0.50 - P_ANGRY[i] * 1 + P_LOVE[i] * 0.75 + P_WOW[i] * 0.50 + P_HAHA[i] * 0.25 + P_THANKFUL[i] * 0.25
    WEIGHT.append(W) # Append into the list 'WEIGHT', the response weight we found of each post 

#for i in range(l_post):
 #   print(WEIGHT[i])

#print(l_post)        
#for i in range(count_post):
 #   print(Post_id[i],R_LIKE[i],R_SAD[i],R_ANGRY[i],R_LOVE[i],R_WOW[i],R_HAHA[i],R_THANKFUL[i], sep = ',')
    

with open('post_weight_response.csv', mode='w') as pw_response_file:
    fieldnames = ['post_id', 'post_likes','post_sad','post_angry','post_love','post_wow','post_haha','post_thankful', 'post_weight']
    writer = csv.DictWriter(pw_response_file, fieldnames=fieldnames)

    writer.writeheader()
    for i in range(l_post):
        writer.writerow({'post_id': post_id[i], 'post_likes': P_LIKE[i] ,'post_sad': P_SAD[i] , 'post_angry': P_ANGRY[i], 'post_love': P_LOVE[i], 'post_wow': P_WOW[i], 'post_haha': P_HAHA[i], 'post_thankful': P_THANKFUL[i], 'post_weight': WEIGHT[i]})
        #print(post_id[i],P_LIKE[i],P_SAD[i],P_ANGRY[i],P_LOVE[i],P_WOW[i],P_HAHA[i],P_THANKFUL[i],WEIGHT[i], sep = ',')
```
