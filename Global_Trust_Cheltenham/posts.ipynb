```python
import csv
import numpy as np

# open the file in universal line ending mode 
with open('post.csv', 'r') as post_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(post_file)
  post = {}
  for row in reader:
    for header, value in row.items():
      try:
        post[header].append(value)
      except KeyError:
        post[header] = [value]
        
#Put all the parameters in separate lists
gid = post['gid']
pid = post['pid']  
uid = post['id']
name = post['name']  
timestamp = post['timeStamp']  
shares = post['shares']  
url = post['url']  
msg = post['msg']  
likes = post['likes'] 

l_uid = len(uid)  # the number of the id's(posts) in the network
#print(l_uid)

count = 0

User = [] 
User_posts = []

S = []
SHARES = []

TOTAL_SHARES = []

x1 = []
x2 = []

message = []

for i in range(l_uid):
    S.append(shares[i]) #Append into the 'S' list, the number of shares for each post
    SHARES = np.asarray(S, dtype=int) #Convert the list 'S' into a numeric array for calculations 
    
for i in range(l_uid):  
    m = 0
    same=1
    X = 0
    
    if msg[i] != '':
        m = m + 1
        
    for k in range(i+1,l_uid):
        if uid[i] == uid[k] and i != k:   #Search for the different user id's in the file
            same = same + 1  #Increases every time we find a cell with the same user_id
            uid[k] = ''
            if msg[k] != '':
                m = m + 1  #Denotes the number of the comments for each post in the network
                
    #print(uid[i], message[i], sep = ',')
    
    if uid[i] != '':   
        count = count + 1  #Denotes the number of different users in the network
        #print(uid[i],same,sep=",")
        User.append(uid[i])  #Append into the list 'User', the id's of the different users we found
        User_posts.append(same)  #Append into the list 'User_posts', the number of the posts each user haw posted in the network
        message.append(m)  #Append into the list 'message', the number of the comments for each post in the network
        #print(uid[i], message[i], sep = ',')
        x1 = SHARES[i] 
        x2 = SHARES[k]
        X = X + (x1 + x2)  #'X' stores temporarily the sum of the shares for each post in the network
        #print(x1)
        #print(x2)
        TOTAL_SHARES.append(X)  #Append into the list 'TOTAL_SHARES', the result of X we found
    

#print(TOTAL_SHARES) 
#print(message)


        # Participation Trust

# AP denotes the average number of posts per user on the Facebook network
# l_uid: the  total number of posts in the network
# count: the  total number of users in the network
AP = l_uid / count
#print(AP)

#Convert the 'User_posts' list into an array of contents-> float numbers, so I can use it for operations(calculations)
USER_POSTS = np.asarray(User_posts, dtype=float)  

# Pu: shows the total number of posts of a user u in the network
# PT: (0 or 1) shows the extent to which a user is involved in uploading posts in the network
Pu = [] * count
PT = [] * count

for i in range(count):
    Pu.append(USER_POSTS[i])
    print(Pu[i])
    print('The user ' + User[i] + ' has posted ' + str(User_posts[i]) + ' posts in total in the network.')
    if Pu[i] >= AP: 
        PTu = 1 
        PT.append(PTu)
        print('The user ' + User[i] + ' is PARTICIPATIVE in the network.')
    else:
        PTu = 0
        PT.append(PTu)
        print('The user ' + User[i] + ' is NON-PARTICIPATIVE in the network.')
        
#print(PT)

with open('user_PT.csv', mode='w') as user_PT_file:
    fieldnames = ['user_id','user_posts','Participation Trust','user_shares', 'user_comments']
    writer = csv.DictWriter(user_PT_file, fieldnames=fieldnames)

    writer.writeheader()
    for i in range(count):
        writer.writerow({'user_id': User[i] ,'user_posts': User_posts[i] , 'Participation Trust': PT[i], 'user_shares': TOTAL_SHARES[i], 'user_comments': message[i]})
        #print(User[i],User_posts[i],PT[i],message[i],sep = ",")
```
