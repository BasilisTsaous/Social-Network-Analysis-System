```python
import csv
import numpy as np

# open the file in universal line ending mode 
with open('post.csv', 'r') as post_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(post_file)
  post = {}
  for row in reader:
    for header, value in row.items():
      try:
        post[header].append(value)
      except KeyError:
        post[header] = [value]
        
#Put all the parameters in separate lists
gid = post['gid']
pid = post['pid']  
uid = post['id']
name = post['name']  
timestamp = post['timeStamp']  
shares = post['shares']  
url = post['url']  
msg = post['msg']  
likes = post['likes'] 

l_uid = len(uid)  # the number of the id's(posts) in the network

P_id = [] 
Posts = []
post_comments = []

count_p = 0

S = []
SHARES = []
post_shares = []

x1 = []
x2 = []

for i in range(l_uid):
    S.append(shares[i])  #Append into the 'S' list, the number of shares for each post
    SHARES = np.asarray(S, dtype=int)  #Convert the list 'S' into a numeric array for calculations 
    

for i in range(l_uid):  
    m = 0
    same_p = 1
    X = 0
    
    if msg[i] != '':
        m = m + 1
        
    for k in range(i+1,l_uid):
        if pid[i] == pid[k] and i != k:  #Search for the different post id's in the file
            same_p = same_p + 1  #Increases every time we find a cell with the same pid(post id)
            pid[k] = ''
            if msg[k] != '':
                m = m + 1  #Denotes the number of the comments for each post(post id) in the network
             
    if pid[i] != '':   
        count_p = count_p + 1  #Denotes the number of different posts in the network
        #print(uid[i],same,sep=",")
        P_id.append(pid[i])  #Append into the list 'P_id', the id's of the different posts we found 
        Posts.append(same_p) #Append into the list 'Posts', the number of the posts, for each post id in the network
        post_comments.append(m)  #Append into the list 'post_comments', the number of the comments for each post(post id) in the network
        
        x1 = SHARES[i] 
        x2 = SHARES[k]
        X = X + (x1 + x2)  #'X' stores temporarily the sum of the shares for each post(post id) in the network
        #print(x1)
        #print(x2)
        post_shares.append(X)  #Append into the list 'post_shares', the result of X we found
    
#print(count_p)
#for i in range(count_p):
 #   print(P_id[i], Posts[i], post_comments[i], post_shares[i], sep =',' )
    

with open('final_post.csv', mode='w') as final_post_file:
    fieldnames = ['post_id', 'number_of_posts', 'post_comments', 'post_shares']
    writer = csv.DictWriter(final_post_file, fieldnames=fieldnames)

    writer.writeheader()
    for i in range(count_p):
        writer.writerow({'post_id': P_id[i], 'number_of_posts': Posts[i] ,'post_comments': post_comments[i] , 'post_shares': post_shares[i]})
        #print print(P_id[i], Posts[i], post_comments[i], post_shares[i], sep =',' ) 
    

      
```
