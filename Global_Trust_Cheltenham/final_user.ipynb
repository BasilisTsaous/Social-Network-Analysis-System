```python
import csv
import numpy as np

# open the file in universal line ending mode 
with open('user_PT.csv', 'r') as user_pt_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(user_pt_file)
  pt = {}
  for row in reader:
    for header, value in row.items():
      try:
        pt[header].append(value)
      except KeyError:
        pt[header] = [value]
        
#Put all the parameters in separate lists
u_id = pt['user_id']
u_posts = pt['user_posts']
u_comments = pt['user_comments']  
u_shares = pt['user_shares']

l_id = len(u_id)  # the number of the id's in the network


# open the file in universal line ending mode 
with open('user_response.csv', 'r') as user_response_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(user_response_file)
  resp = {}
  for row in reader:
    for header, value in row.items():
      try:
        resp[header].append(value)
      except KeyError:
        resp[header] = [value]

#Put all the parameters in separate lists   
user_id = resp['user_id']
user_likes = resp['user_likes']
user_sad = resp['user_sad']
user_angry = resp['user_angry']
user_love = resp['user_love']
user_wow = resp['user_wow']
user_haha = resp['user_haha']
user_thankful = resp['user_thankful']

lu = len(user_id)  # the number of the id's in the network

USER = []
POSTS = []
COMMENTS = []
SHARES = []
LIKES = []
SAD = []
ANGRY = []
LOVE = []
WOW = []
HAHA = []
THANKFUL = []

count = 0
for i in range(l_id):
    for k in range(lu):
        if u_id[i] == user_id[k]:
            count = count + 1
            
            USER.append(u_id[i])
            POSTS.append(u_posts[i])
            COMMENTS.append(u_comments[i])
            SHARES.append(u_shares[i])
            LIKES.append(user_likes[k])
            SAD.append(user_sad[k])
            ANGRY.append(user_angry[k])
            LOVE.append(user_love[k])
            WOW.append(user_wow[k])
            HAHA.append(user_haha[k])
            THANKFUL.append(user_thankful[k])

#print(count)  

with open('final_user.csv', mode='w') as final_file:
    fieldnames = ['user_id','user_posts','user_comments','user_shares','user_likes', 'user_sad','user_love', 'user_angry','user_wow','user_haha','user_thankful']
    writer = csv.DictWriter(final_file, fieldnames=fieldnames)

    writer.writeheader()
    for i in range(count):
        writer.writerow({'user_id': USER[i] ,'user_posts': POSTS[i],'user_comments': COMMENTS[i] , 'user_shares': SHARES[i] , 'user_likes': LIKES[i], 'user_sad': SAD[i], 'user_love': LOVE[i] ,'user_angry': ANGRY[i] ,'user_wow': WOW[i], 'user_haha': HAHA[i] ,'user_thankful': THANKFUL[i]})
```
