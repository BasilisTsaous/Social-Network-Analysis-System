```python
import csv
import numpy as np

# open the file in universal line ending mode 
with open('post.csv', 'r') as post_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(post_file)
  post = {}
  for row in reader:
    for header, value in row.items():
      try:
        post[header].append(value)
      except KeyError:
        post[header] = [value]
        
#Put all the parameters in separate lists
gid = post['gid']
pid = post['pid']  
uid = post['id']
name = post['name']  
timestamp = post['timeStamp']  
shares = post['shares']  
url = post['url']  
msg = post['msg']  
likes = post['likes'] 

l_pid = len(pid)  # the number of the id's(posts) in the network


# open the file in universal line ending mode 
with open('post_weight_response.csv', 'r') as pw_response_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(pw_response_file)
  pw = {}
  for row in reader:
    for header, value in row.items():
      try:
        pw[header].append(value)
      except KeyError:
        pw[header] = [value]

#Put all the parameters in separate lists       
post_id = pw['post_id']
post_likes = pw['post_likes']
post_sad = pw['post_sad']
post_angry = pw['post_angry']
post_love = pw['post_love']
post_wow = pw['post_wow']
post_haha = pw['post_haha']
post_thankful = pw['post_thankful']
post_weight = pw['post_weight']

l_post = len(post_id)  # the number of the post id's in the network


P_LIKES = np.asarray(post_likes, dtype = int)
P_SAD = np.asarray(post_sad, dtype = int)
P_LOVE = np.asarray(post_love, dtype = int)
P_ANGRY = np.asarray(post_angry, dtype = int)
P_WOW = np.asarray(post_wow, dtype = int)
P_HAHA = np.asarray(post_haha, dtype = int)
P_THANKFUL = np.asarray(post_thankful, dtype = int)

S1 = []
S2 = []
S3 = []
S4 = []
S5 = []
S6 = []
S7 = []

user = []

count = 0

for i in range(l_pid):
    s1 = 0
    s2 = 0
    s3 = 0
    s4 = 0
    s5 = 0
    s6 = 0
    s7 = 0
    for k in range(l_post):
        if pid[i] == post_id[k]:  #Search for the posts(post id's) that exist in both files
            count = count + 1
            
            s1 = s1 + P_LIKES[k]
            s2 = s2 + P_SAD[k]
            s3 = s3 + P_LOVE[k]
            s4 = s4 + P_ANGRY[k]
            s5 = s5 + P_WOW[k]
            s6 = s6 + P_HAHA[k]
            s7 = s7 + P_THANKFUL[k]
            
    S1.append(s1)
    S2.append(s2) 
    S3.append(s3) 
    S4.append(s4) 
    S5.append(s5) 
    S6.append(s6)
    S7.append(s7) 
    user.append(uid[i])

#print(count)    
#for i in range(count):
 #   print(user[i],pid[i],S1[i], sep = ',')
    
p_S1 = np.asarray(S1, dtype = int)
p_S2 = np.asarray(S2, dtype = int)
p_S3 = np.asarray(S3, dtype = int)
p_S4 = np.asarray(S4, dtype = int)
p_S5 = np.asarray(S5, dtype = int)
p_S6 = np.asarray(S6, dtype = int)
p_S7 = np.asarray(S7, dtype = int)    

P_S1 = []
P_S2 = []
P_S3 = []
P_S4 = []
P_S5 = []
P_S6 = []
P_S7 = []
USER = []
    

c = 0
for i in range(count):
    if user[i] != '':
        sum1 = p_S1[i]
        sum2 = p_S2[i]
        sum3 = p_S3[i]
        sum4 = p_S4[i]
        sum5 = p_S5[i]
        sum6 = p_S6[i]
        sum7 = p_S7[i]
    
        for k in range(i+1,count):
            if user[i] == user[k] and i != k:   #Search for the different user id's in the file
                user[k] = ''

                sum1 = sum1 + p_S1[k]
                sum2 = sum2 + p_S2[k]
                sum3 = sum3 + p_S3[k]
                sum4 = sum4 + p_S4[k]
                sum5 = sum5 + p_S5[k]
                sum6 = sum6 + p_S6[k]
                sum7 = sum7 + p_S7[k]

        if user[i] != '':
            c = c + 1

            P_S1.append(sum1)
            P_S2.append(sum2)
            P_S3.append(sum3)
            P_S4.append(sum4)
            P_S5.append(sum5)
            P_S6.append(sum6)
            P_S7.append(sum7)
            USER.append(user[i])

        
#print(c)

with open('user_response.csv', mode='w') as user_resp_file:
    fieldnames = ['user_id','user_likes', 'user_sad','user_love', 'user_angry','user_wow','user_haha','user_thankful']
    writer = csv.DictWriter(user_resp_file, fieldnames=fieldnames)

    writer.writeheader()
    for i in range(c):
        writer.writerow({'user_id': USER[i] , 'user_likes': P_S1[i], 'user_sad': P_S2[i], 'user_love': P_S3[i] ,'user_angry': P_S4[i] ,'user_wow': P_S5[i], 'user_haha': P_S6[i] ,'user_thankful': P_S7[i]})
```
