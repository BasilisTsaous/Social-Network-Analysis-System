```python
import csv
import numpy as np
#import pandas

# open the file in universal line ending mode 
with open('INDIRECT_TRUST.csv', 'r') as it_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(it_file)
  it = {}
  for row in reader:
    for header, value in row.items():
      try:
        it[header].append(value)
      except KeyError:
        it[header] = [value]
        
#Put all the parameters in separate lists 
u_id = it['user_id']
u_posts = it['user_posts']
p_trust = it['Participation Trust']
r_trust = it['Response Trust']
i_trust = it['Indirect Trust']

l = len(u_id)  #the number of the id's in the network


# open the file in universal line ending mode 
with open('final_user.csv', 'r') as final_user_file:
  # read the file as a dictionary for each row ({header : value})
  reader = csv.DictReader(final_user_file)
  f = {}
  for row in reader:
    for header, value in row.items():
      try:
        f[header].append(value)
      except KeyError:
        f[header] = [value]

#Put all the parameters in separate lists   
user_id = f['user_id']
user_posts = f['user_posts']
user_comments = f['user_comments']
user_shares = f['user_shares']
user_likes = f['user_likes']
user_sad = f['user_sad']
user_angry = f['user_angry']
user_love = f['user_love']
user_wow = f['user_wow']
user_haha = f['user_haha']
user_thankful = f['user_thankful']

lu = len(user_id)  # the number of the id's in the network


comments = np.asarray(user_comments, dtype=int)
shares = np.asarray(user_shares, dtype=int)
likes = np.asarray(user_likes, dtype=int)
sad = np.asarray(user_sad, dtype=int)
angry = np.asarray(user_angry, dtype=int)
love = np.asarray(user_love, dtype=int)
wow = np.asarray(user_wow, dtype=int)
haha = np.asarray(user_haha, dtype=int)
thankful = np.asarray(user_thankful, dtype=int)

TRUST = []
P_TRUST = []
R_TRUST = []
USER = []
POSTS = []
COMMENTS = []
SHARES = []
LIKES = []
SAD = []
ANGRY = []
LOVE = []
WOW = []
HAHA = []
THANKFUL = []

s = []
S = []

count = 0

for i in range(l):
    for k in range(lu):
        if u_id[i] == user_id[k]:
            count = count + 1
            
            s = comments[i] + shares[i] + likes[i] + sad[i] + angry[i] + love[i] + wow[i] + haha[i] + thankful[i]
            S.append(s)
            
            TRUST.append(i_trust[i])
            P_TRUST.append(p_trust[i])
            R_TRUST.append(r_trust[i])
            USER.append(user_id[k])
            POSTS.append(user_posts[k])
            COMMENTS.append(user_comments[k])
            SHARES.append(user_shares[k])
            LIKES.append(user_likes[k])
            SAD.append(user_sad[k])
            ANGRY.append(user_angry[k])
            LOVE.append(user_love[k])
            WOW.append(user_wow[k])
            HAHA.append(user_haha[k])
            THANKFUL.append(user_thankful[k])


#print(count)

with open('linear_data.csv', mode='w') as data_file:
    fieldnames = ['user_id','user_posts','user_comments','user_shares','user_likes', 'user_sad','user_love','user_angry','user_wow','user_haha','user_thankful','total responses','PARTICIPATION_TRUST','RESPONSE_TRUST','TRUST']
    writer = csv.DictWriter(data_file, fieldnames=fieldnames)

    writer.writeheader()
    for i in range(count):
        writer.writerow({'user_id': USER[i] ,'user_posts': POSTS[i],'user_comments': COMMENTS[i] , 'user_shares': SHARES[i] , 'user_likes': LIKES[i], 'user_sad': SAD[i], 'user_love': LOVE[i] ,'user_angry': ANGRY[i] ,'user_wow': WOW[i], 'user_haha': HAHA[i] ,'user_thankful': THANKFUL[i], 'total responses': S[i], 'PARTICIPATION_TRUST': P_TRUST[i], 'RESPONSE_TRUST': R_TRUST[i],'TRUST': TRUST[i]})
```
