```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('INDIRECT_TRUST.csv')

plt.hist(data['Indirect Trust'], range = [0.1,0.8])
plt.title('Indirect Trust Histogram')
plt.xlabel('IT Trust')
plt.ylabel('Number of Users')
plt.show()
```
