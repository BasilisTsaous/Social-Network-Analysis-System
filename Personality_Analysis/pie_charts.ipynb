```python
import pandas as pd
import matplotlib.pyplot as plt

f1 = pd.read_csv('user_extroversion.csv')

df1 = f1.groupby(['SCORE']).count()['user_id']

df1.plot(kind='pie', subplots=True, shadow = True, startangle=90, figsize=(6,6), autopct='%1.1f%%')

plt.show()
```
